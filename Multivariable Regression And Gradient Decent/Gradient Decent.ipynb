We can perform generic Gradient Decent , batch gradient decent stochastic gradient decent and mini gradient decent by doing some changes in step gradient
### generic Gradient Decent
in this we have to pass array of m and run loop for updating all m 
### batch gradient decent
in this we have to update m c every time when we are updating m_slope and c_slope value 

```python
def step_gradient(point,lr,m,c):
    m_slope=0;
    c_slope=0;
    M=len(point)
    for i in range(M):
        x=point[i,0]
        y=point[i,1]
        m_slope+=(-2/M)*(y-m*x-c)*x
        c_slope+=(-2/M)*(y-m*x-c)
    m_val=m-m_slope*lr
    c_val=c-c_slope*lr
    return m_val,c_val

def gd(point,learning_rate,number_iteration):
    m=0
    c=0
    for i in range(number_iteration):
        m,c=step_gradient(point,learning_rate,m,c)
        print(i,"cost ",cost(point,m,c))
    return m,c
```

```python
def cost(point,m,c):
    total_cost=0
    M=len(point)
    for i in range(M):
        x=point[i,0]
        y=point[i,1]
        total_cost+=(1/M)*((y-m*x-c)**2)
        return total_cost
```
